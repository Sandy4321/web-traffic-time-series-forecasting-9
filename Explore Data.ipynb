```python
# from https://github.com/thepaulm/pyplot_examples/blob/master/Pyplot%20Examples.ipynb
def pta(*args, color_rotate=True):
    '''pta = plot them all'''
    count = len(args)
    w = 20
    h = 4 * count
    fig = plt.figure(1, figsize=(w, h))
    for i, a in enumerate(args):
        ax = fig.add_subplot(count, 1, i+1)
        if color_rotate:
            for _ in range(i):
                ax.plot([],[])
        ax.plot(a)
```

```python
def grab_data(df, count):
    datas = []
    for a in range(count):
        ri = np.random.randint(0, df.shape[0])
        npd = df[ri:ri+1].fillna(0).as_matrix()
        npd = npd.squeeze()[1:].astype('float32')
        datas.append(npd)
    return datas
```

```python
def shaper(data, timesteps, predictions):
    outdata = []
    for a in range(0, len(data) - (timesteps * predictions)):
        d = data[a:a + timesteps * predictions]
        d = d.reshape((timesteps, predictions))
        outdata.append(d)
    return np.array(outdata)
```

```python
d = shaper(npdata, 1, 60)
```
